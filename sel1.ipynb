```python
job_title = "Data Analyst"

# Find all positions of 'a'
indices = [i for i, c in enumerate(job_title) if c == 'a']

# Replace only the 2nd one (index 1 â†’ second occurrence)
if len(indices) >= 2:
    pos = indices[2]
    job_title = job_title[:pos] + "o" + job_title[pos+1:]

print(job_title)
```

```python
job_title = "Data Analyst"

pos = job_title.rfind("a")
if pos != -1:
    job_title = job_title[:pos] + "o" + job_title[pos+1:]

print(job_title)
```

```python
import re

job_title = "Data Analyst"

def replace_nth(s, sub, repl, n):
    matches = [m.start() for m in re.finditer(sub, s)]
    if len(matches) >= n:
        i = matches[n-1]
        s = s[:i] + repl + s[i+len(sub):]
    return s

print(replace_nth(job_title, "a", "o", 2))  # Replace 2nd 'a'
```
